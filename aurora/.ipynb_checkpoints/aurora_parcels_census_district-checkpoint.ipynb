```python
import pandas as pd
```

```python
# Write out the parcel IDs and addresses to upload for geocoding using:
# https://geocoding.geo.census.gov/geocoder/geographies/addressbatch?form
# Choose "Census2010_Current" as these tract IDs match the CEJST data
address_df = pd.read_csv(r"C:/Users/ltaylor2/NREL/Building Performance Standards TA - Equity Analysis/Data Analysis/CoStar Data/PropertyAndTenantDataCombined.csv")
address_df_filtered = address_df[['Address','City', 'State', 'Zip']]
print(address_df_filtered)
address_df_filtered.to_csv(r"C:\Users\ltaylor2\Documents\BPS Charts\aurora_addresses_for_geocoding.csv",header=False)
```

```python
# After manually uploading the geocoded addresses,
# read the returned file. Column names per census.gov documentation.
geocode_cols = [
    'NREL_ID',
    'Input Address',
    'TIGER Address Range Match Indicator',
    'TIGER Match Type',
    'TIGER Output Address',
    'LongitudeAndLatitude',
    'Tigerline ID',
    'Tigerline ID Side',
    'State',
    'County',
    'Tract',
    'Block'
]

geo = pd.read_csv(r"C:\Users\ltaylor2\Documents\BPS Charts\GeocodeResults.csv", names=geocode_cols, dtype=object)
geo['NREL_ID'] = geo['NREL_ID'].astype('int64')
geo.set_index('NREL_ID', inplace=True)

# Check the number of projects with matches vs. no matches
print('Geocoding address matching results:')
print(geo['TIGER Address Range Match Indicator'].value_counts())

# Save the unmatched addresses to CSV for investigation
geo[~(geo['TIGER Address Range Match Indicator'] == 'Match')].to_csv(r"C:\Users\ltaylor2\Documents\BPS Charts\'unmatched_parcel_addresses.csv")

# Limit the geocode to just 
geo = geo[geo['TIGER Address Range Match Indicator'] == 'Match']
geo
```

```python
# Join the geographic data onto the parcel data by NREL ID
nonres = pd.merge(address_df_filtered, geo, left_index=True, right_index=True)
nonres.head()
```

```python
# Strip the 0 from the front of state_y values
nonres['State_y'] = nonres ['State_y'].str.lstrip('0')
print(nonres['State_y'])

# Combine the State, County, and Tract into a single field
def census_tract_id(row):
    # STATE+COUNTY+TRACT
    # 2+3+6=11
    state = row['State_y']
    county = row['County']
    tract = row['Tract']
    gisjoin = f'{state}{county}{tract}'

    if not (len(gisjoin) == 10 or len(gisjoin) == 11):
        print(f'Failed to make census ID for state:{state}, county:{county}, tract:{tract}')
        return pd.NA
        
    # assert(len(gisjoin) == 11)

    return gisjoin

nonres['census_tract_id'] = nonres.apply(lambda row: census_tract_id(row), axis=1)
nonres['census_tract_id'].value_counts()
```

```python
# Load the EEEJ
cejst = pd.read_csv(r"C:\Users\ltaylor2\Documents\BPS Charts\1.0-communities.csv", low_memory=False, dtype=object)
cejst
```

```python
# Merge CEJST data on census tract ID
bef = len(nonres)
print(f'Before joining CEJST on geographies, Nonres had {bef} entries')

nonres = pd.merge(nonres, cejst, left_on='census_tract_id', right_on='Census tract 2010 ID', how='inner')

aft = len(nonres)
print(f'After joining CEJST on census geographies, Nonres had {aft} entries')
nomatch = bef - aft
if nomatch > 0:
    print(f'This means there were {nomatch} census geographies not matched. If this number is high, make sure you choose "Census2010_current on the geocoding site.')

nonres.index.name = 'NREL_ID'

nonres
```

```python
# nonres3 = nonres[['census_tract_id', 'Census tract ID']]
nonres[nonres['Census tract 2010 ID'].isna()]['census_tract_id'].value_counts()
nonres = nonres[['Address', 'Identified as disadvantaged']]
nonres.to_csv(r"C:\Users\ltaylor2\Documents\BPS Charts\DisadvantagedStatusOfBuildingsAurora.csv")
```

```python
# Put each building into a bin by floor area
def size_bin(row):
    sf = row['COM_GROSS_AREA']
    if sf < 1_000:
        b = '_1000'
    elif sf < 5_000:
        b = '1001_5000'
    elif sf < 10_000:
        b = '5001_10000'
    elif sf < 25_000:
        b = '10001_25000'
    elif sf < 50_000:
        b = '25001_50000'
    elif sf < 100_000:
        b = '50001_100000'
    elif sf < 200_000:
        b = '100001_200000'
    elif sf < 500_000:
        b = '200001_500000'
    elif sf < 1_000_000:
        b = '500001_1000000'
    else:
        b = 'over_1mil'

    return b

nonres['floor_area_category'] = nonres.apply(lambda row: size_bin(row), axis=1)
```

```python
# Put each building into a bin by floor area
def ownership_type(row):
    name = row['OWNER_NAME'].lower()
    if 'inv' in name:
        return 'investor'
    else:
        return 'non-investor'

nonres['owner_type'] = nonres.apply(lambda row: ownership_type(row), axis=1)
```
